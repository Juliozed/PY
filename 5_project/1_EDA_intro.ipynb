```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

filter both 

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')].copy()
```

explore job locations, home office, degree req, health insurance? then companies

start with value_counts() method

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Analyst in the US')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

home office?  

```python
dict_column = {
    'job_work_from_home': 'Home Office',
    'job_no_degree_mention': 'Degree Req.',
    'job_health_insurance': 'Health Insurance Offered '
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches(12, 5)

# now our for loop
# we must put . our key and value in the dictionary in a tuple for it to loop! 
for i, (column, title) in enumerate(dict_column.items()):
    print(i, column, title)
    ax[i].pie(df_DA_US[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False', 'True']) # must be in correct order!
    ax[i].set_title(title)

plt.show()
```

now company name

```python
df_plot = df_DA_US['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Companies Hiring Data Analyst in the US')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
